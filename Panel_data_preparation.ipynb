```python
import numpy as np
import pandas as pd
```

```python
neet_rate_full = pd.read_csv('./data/education_system/NEET_men_women_altogether.csv')
exp_lmp_full = pd.read_csv('./data/labour_market/public_exp_LMP.csv')
student_teacher_ratio_full = pd.read_csv('./data/education_system/student_teacher_ratio.csv')
min_wage_full = pd.read_csv('./data/labour_market/real minimum wage from 2001 to 2019.csv')
```

```python
years_schooling_full = pd.read_csv('./data/education_system/mean-years-of-schooling-world.csv')
avg_class_size_full = pd.read_csv('./data/education_system/Avg_class_size_primary_education.csv')
spending_educ_full = pd.read_csv('./data/education_system/spendings_in_education.csv')
education_level_and_parents_full = pd.read_csv('./data/education_system/education_level_depending_on_those_of_the_parents_2012_2015.csv')
```

```python
avg_duration_unemployment_full = pd.read_csv('./data/labour_market/average duration of unemployment.csv')
strictness_emply_protec_full = pd.read_csv('./data/labour_market/strictness of employment protection from 1998 to 2018.csv')
short_time_workers_full = pd.read_csv('./data/labour_market/percentage of short time workers from 2000.csv')
involuntary_pt_workers_full = pd.read_csv('./data/labour_market/percentage of involuntary part time workers from 2000.csv')
well_matched_skills_jobs_full = pd.read_csv('./data/labour_market/percentage of well matched skills for jobs.csv', dtype='unicode')
ft_and_pt_employ_full = pd.read_csv('./data/labour_market/percentage of full time part time employment.csv')
marginally_attached_workers_full = pd.read_csv('./data/labour_market/percentage of marginally attached workers from 2000.csv')
employees_bargain_full = pd.read_csv('./data/labour_market/percentage of employees that can bargain from 1998.csv')
```

```python
oecd_countries = {'AUS': 'Australia', 'AUT': 'Austria', 'BEL': 'Belgium', 'CAN': 'Canada', 'CHL': 'Chile', 'COL': 'Colombia', 'CZE': 'Czech Republic', 'DNK': 'Denmark', 'EST': 'Estonia', 'FIN': 'Finland', 'FRA': 'France', 'DEU': 'Germany', 'GRC': 'Greece', 'HUN': 'Hungary', 'ISL': 'Iceland', 'IRL': 'Ireland', 'ISR': 'Israel', 'ITA': 'Italy', 'JPN': 'Japan', 'KOR': 'Korea', 'LVA': 'Latvia', 'LTU': 'Lithuania', 'LUX': 'Luxembourg', 'MEX': 'Mexico', 'NLD': 'Netherlands', 'NZL': 'New Zealand', 'NOR': 'Norway', 'POL': 'Poland', 'PRT': 'Portugal', 'SVK': 'Slovakia', 'SVN': 'Slovenia', 'ESP': 'Spain', 'SWE': 'Sweden', 'CHE': 'Switzerland', 'TUR': 'Turkey', 'GBR': 'United Kingdom', 'USA': 'United States'}
code_countries = [code for code in oecd_countries.keys()]
```

# Create panel data set for the NEET rate and the 3 explanatory variables

## NEET rate

```python
neet_rate = neet_rate_full[['LOCATION', 'SUBJECT', 'TIME', 'Value']]
neet_rate_oecd = pd.DataFrame(columns=neet_rate.columns)
for code in code_countries:
    neet_rate_oecd = neet_rate_oecd.append(neet_rate[neet_rate.LOCATION==code], ignore_index=True)
neet_rate_oecd.columns = ['Country', 'Subject', 'Time', 'NEET']
```

```python
subject = '15_29'
neet_rate_oecd = neet_rate_oecd[neet_rate_oecd.Subject == subject].drop(['Subject'], axis=1).reset_index(drop=True)
```

## Student-Teacher ratio

```python
student_teacher_ratio = student_teacher_ratio_full[['COUNTRY', 'ISC11_LEVEL_CAT', 'YEAR', 'Value']]
student_teacher_ratio.columns = ['Country', 'ISC11_LEVEL_CAT', 'Time', 'STR']
```

```python
student_teacher_ratio_l0=student_teacher_ratio[student_teacher_ratio.ISC11_LEVEL_CAT=='L0'].drop(['ISC11_LEVEL_CAT'], axis=1).reset_index(drop=True)
student_teacher_ratio_l1=student_teacher_ratio[student_teacher_ratio.ISC11_LEVEL_CAT=='L1'].drop(['ISC11_LEVEL_CAT'], axis=1).reset_index(drop=True)
student_teacher_ratio_l2=student_teacher_ratio[student_teacher_ratio.ISC11_LEVEL_CAT=='L2'].drop(['ISC11_LEVEL_CAT'], axis=1).reset_index(drop=True)
student_teacher_ratio_l3=student_teacher_ratio[student_teacher_ratio.ISC11_LEVEL_CAT=='L3'].drop(['ISC11_LEVEL_CAT'], axis=1).reset_index(drop=True)
student_teacher_ratio_l5t8=student_teacher_ratio[student_teacher_ratio.ISC11_LEVEL_CAT=='L5T8'].drop(['ISC11_LEVEL_CAT'], axis=1).reset_index(drop=True)
```

```python
#Let's create a STR df corresponding to the secondary eductation system
student_teacher_ratio_l2_l3=student_teacher_ratio[(student_teacher_ratio.ISC11_LEVEL_CAT=='L2')|(student_teacher_ratio.ISC11_LEVEL_CAT=="L3")].drop(['ISC11_LEVEL_CAT'], axis=1).reset_index(drop=True)
```

## Public expenditure on LMP

```python
exp_lmp = exp_lmp_full[['LFS_COUNTRY', 'PROG', 'Time', 'Value']]
exp_lmp_oecd = exp_lmp[exp_lmp.PROG==100].drop(['PROG'], axis=1).reset_index(drop=True)
exp_lmp_oecd.columns = ['Country', 'Time', 'Exp_LMP']
```

## Minimum wage

```python
min_wage = min_wage_full[['COUNTRY','SERIES','PERIOD','TIME','Value']]
min_wage = min_wage[(min_wage.SERIES == 'PPP') & (min_wage.PERIOD == 'A')].drop(['SERIES', 'PERIOD'], axis=1).reset_index(drop=True)
min_wage_oecd = pd.DataFrame(columns=min_wage.columns)
for code in code_countries:
    min_wage_oecd = min_wage_oecd.append(min_wage[min_wage.COUNTRY == code],ignore_index=True)
min_wage_oecd.columns = ['Country', 'Time', 'Min_Wage']
```

## Concatenate all variables in one df

```python
df_model_l1 = pd.DataFrame(columns=['Country', 'Time', 'NEET', 'Exp_LMP', 'STR', 'Min_Wage'])
years = [year for year in range(2013,2019,1)]
for country in code_countries:
    neet_rate_country = neet_rate_oecd[(neet_rate_oecd.Country==country)&(neet_rate_oecd.Time.isin(years))]
    exp_lmp_country = exp_lmp_oecd[(exp_lmp_oecd.Country==country)&(exp_lmp_oecd.Time.isin(years))]
    student_teacher_ratio_country = student_teacher_ratio_l1[(student_teacher_ratio_l1.Country==country)&(student_teacher_ratio_l1.Time.isin(years))]
    min_wage_country = min_wage_oecd[(min_wage_oecd.Country==country)&(min_wage_oecd.Time.isin(years))]
    for year in years:
        neet = neet_rate_country[neet_rate_country.Time==year]['NEET'].values
        exp_lmp_value = exp_lmp_country[exp_lmp_country.Time==year]['Exp_LMP'].values
        str_value = student_teacher_ratio_country[student_teacher_ratio_country.Time==year]['STR'].values
        min_wage_value = min_wage_country[min_wage_country.Time==year]['Min_Wage'].values
        if (neet.size>0) and (exp_lmp_value.size>0) and (str_value.size>0) and (min_wage_value.size>0):
            if (neet==neet) and (exp_lmp_value==exp_lmp_value) and (str_value==str_value) and (min_wage_value==min_wage_value):
                df_model_l1 = df_model_l1.append({'Country': country, 'Time': year, 'NEET': neet[0], 'Exp_LMP': exp_lmp_value[0], 'STR': str_value[0], 'Min_Wage': min_wage_value[0]}, ignore_index=True)
```

```python
df_model_l1.to_csv('./panel_data_l1.csv', index=False)
```

```python
#Creating the dataframe for the education level l2 and l3 combine
df_model_l2_l3 = pd.DataFrame(columns=['Country','Time','NEET','Exp_LMP','STR'])
years = [year for year in range(2013,2019,1)]
for country in code_countries:
    neet_rate_country = neet_rate_oecd[(neet_rate_oecd.Country==country)&(neet_rate_oecd.Time.isin(years))]
    exp_lmp_country = exp_lmp_oecd[(exp_lmp_oecd.Country==country)&(exp_lmp_oecd.Time.isin(years))]
    student_teacher_ratio_country = student_teacher_ratio_l2_l3[(student_teacher_ratio_l2_l3.Country==country)&(student_teacher_ratio_l2_l3.Time.isin(years))]
    for year in years:
        neet = neet_rate_country[neet_rate_country.Time==year]['NEET'].values
        exp_lmp_value = exp_lmp_country[exp_lmp_country.Time==year]['Exp_LMP'].values
        str_value = student_teacher_ratio_country[student_teacher_ratio_country.Time==year]['STR'].values
        str_value=str_value[~np.isnan(str_value)] #Suppressing nan
        if (neet.size>0) and (exp_lmp_value.size>0) and (str_value.size>0):
            if str_value.size==2:
                str_value=np.array([str_value.mean()])
            if (neet==neet) and (exp_lmp_value==exp_lmp_value) and (str_value==str_value):
                df_model_l2_l3 = df_model_l2_l3.append({'Country': country, 'Time': year, 'NEET': neet[0], 'Exp_LMP': exp_lmp_value[0], 'STR': str_value[0]}, ignore_index=True)
```

```python
df_model_l2_l3.to_csv('./panel_data_l2_l3.csv',index = False)
```

## Concatenate all variables in one df

```python
df_educ_l1 = pd.DataFrame(columns=['Country', 'Time', 'Years_schooling', 'Avg_class_size', 'Exp_educ'])
years = [year for year in range(2013,2018)]
for country in code_countries:
    years_sch_country = years_schooling_oecd[(years_schooling_oecd.Country==country)&(years_schooling_oecd.Time.isin(years))]
    avg_class_country = avg_class_size_oecd_l1[(avg_class_size_oecd_l1.Country==country)&(avg_class_size_oecd_l1.Time.isin(years))]
    spd_educ_country = spending_educ_oecd_l1[(spending_educ_oecd_l1.Country==country)&(spending_educ_oecd_l1.Time.isin(years))]
    for year in years:
        schooling = years_sch_country[years_sch_country.Time==year]['Years_schooling'].values
        class_size = avg_class_country[avg_class_country.Time==year]['Avg_class_size'].values
        spd_educ = spd_educ_country[spd_educ_country.Time==year]['Exp_educ'].values
        if (schooling.size>0) and (class_size.size>0) and (spd_educ.size>0):
            if (schooling==schooling) and (class_size==class_size) and (spd_educ==spd_educ):
                df_educ_l1 = df_educ_l1.append({'Country': country, 'Time': year, 'Years_schooling': schooling[0], 'Avg_class_size': class_size[0], 'Exp_educ': spd_educ[0]}, ignore_index=True)
```
